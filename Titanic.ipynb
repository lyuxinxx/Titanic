```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('train.csv', index_col = 'PassengerId', na_values = ['NULL'])
data
```

```python
nacounts = np.sum(data.isnull())
nacounts = pd.DataFrame({"Count" : nacounts, "Percentage" : nacounts/891})
nacounts
```

```python
continuousData = data.loc[:,["Age","Fare"]]
continuousData.describe()
```

```python
discreteData = data.loc[:,['Survived','Pclass','Sex','SibSp','Parch','Embarked']]
discreteData.groupby('Survived').size().plot.bar(color='k', alpha=0.7,rot=0)
plt.show()
```

```python
discreteData.groupby('Pclass').size().plot.bar(color='b', alpha=0.5,rot=0)
plt.show()
```

```python
discreteData.groupby('Sex').size().plot.bar(color='r', alpha=0.7,rot=0)
plt.show()
```

```python
discreteData.groupby('SibSp').size().plot.bar(color='y', alpha=0.7,rot=0)
plt.show()
```

```python
discreteData.groupby('Parch').size().plot.bar(color='g', alpha=0.7,rot=0)
plt.show()
```

```python
discreteData.groupby('Embarked').size().plot.bar(color='b', alpha=0.7,rot=0)
plt.show()
```
